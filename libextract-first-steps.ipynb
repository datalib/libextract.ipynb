So to begin, let's import our libraries for this brief "data science" exercise:

```python
# for those using ipython notebook
%matplotlib inline 
import re # regex
import pandas as pd 
from requests import get

from libextract import extract # our standard extraction method
from libextract.strategies import TABULAR # for when dealing with tables
from libextract import prototypes # currently found in 'fuzzy-table-formatter' branch
```

... let's download our html

```python
url = "http://docquery.fec.gov/cgi-bin/forms/C00574624/1004025/"
tedcruz = get(url)
```

Normally at this point many people will have to ``Alt-Tab`` from browser to shell/editor in a phase data scientists/analysts/visualizers sometimes refer to as *[data wrangling/munching](http://en.wikipedia.org/wiki/Data_wrangling)*. 

This munging step is usually quite involved - there are wrongly formatted ``<table>``'s; there non-``<table>`` yet still *tabular* data, and so on.

For an idea of how a data munging step plays out, visit [Chris Moffitt's post on web scraping and visualizing](http://pbpython.com/web-scraping-mn-budget.html).

---

Now here is libextract doing some of the data munging for you:

```python
# convert_table is something not yet in the master branch
strat = TABULAR + (prototypes.convert_table,) 
# top 5 predicted html elements/dicts of tabular data
tabs = list(extract(tedcruz.content, strategy=strat)) 
```

Now let's see what libextract extraction algo. returns to you:

```python
for tab in tabs:
    print(type(tab))
```

It's just a list of dictionaries and html elements...

B/c this post was inspired by Willis's tweet, let's use the table he tweeted about (the third ``dict``); let's convert that ``dict`` to a [DataFrame](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.html#pandas.DataFrame)

```python
df1 = pd.DataFrame.from_dict(tabs[2])
df1[:10]
```

Will you look at that! It's practically magic! 

Note: if you'd like to see libextract self-deprecate, run: ``df1 = pd.DataFrame.from_dict(tabs[0])``

Before we start visualizing, let's apply some string-to-numerical magic:

```python
# this function will be "applied" to every element in the column we designate
def convert_num(n): 
    return float(n)

df1['1'] = df1['1'].apply(convert_num)
```

Let's also set the first column as our "index", and then sort:

```python
df1.columns = ['Summary of Expenditures','Dollars']
df1 = df1.set_index('Summary of Expenditures')
df1 = df1.sort('Dollars')
df1 = df1[(df1.index != "8. Subtotal (6 + 7)")]
```

And now let's visualize!

```python
pd.options.display.mpl_style = 'default' # to set a more neutral 'blue'
# Understandably, we will have to set the title by hand, for now... 
df1.plot(kind='barh', figsize=[20, 10],title="Covering Period 01/01/2015 Through 03/31/2015")
```

## Part 2 - List of countries by life expectancy (wikipedia)

What did you find unnatractive in the last section (besides the underwhelming chart)? 

If it's something that you think we can improve within libextract, [please let us know](https://github.com/datalib/libextract/issues)

Now let's repeat the same code from above:

```python
url = "http://en.wikipedia.org/wiki/List_of_countries_by_life_expectancy"
life_expectancy = get(url)

# convert_table is something not yet in the master branch
strat = TABULAR + (prototypes.convert_table,) 
# top 5 predicted html elements/dicts of tabular data
tabs = list(extract(life_expectancy.content, strategy=strat)) 

df1 = pd.DataFrame.from_dict(tabs[0])

# Someday... Someday we will not need to do any data munging!
def convert_num(n): 
    n = re.sub(r'(\[\d*\])','',n)
    #print(n)
    if n: 
        return float(n)
    else: 
        return None
    
df1 = df1.set_index('Entity')

df1['Female life expectancy at birth'] = df1['Female life expectancy at birth'].apply(convert_num)
df1['Male life expectancy at birth'] = df1['Male life expectancy at birth'].apply(convert_num)
df1['Overall life expectancy at birth'] = df1['Overall life expectancy at birth'].apply(convert_num)
```

```python
df1['Female life expectancy at birth']
```

```python
df1 = df1.sort(columns='Overall life expectancy at birth', ascending=True)

pd.options.display.mpl_style = 'default' # to set a more neutral 'blue'
# Understandably, we will have to set the title by hand, for now... 
ax = df1.plot(kind='barh', figsize=[20,100],
              xticks=list(range(101))[0::5],
              xlim=(30,95))
```

```python
# grrr, there's more visualization code than in libextract's codebase... 
# jk, but this part is annoying
title="List of countries by life expectancy"
ax.xaxis.tick_top()

ax.text(0.5, 1.0125, title,
         horizontalalignment='center',
         fontsize=20,
         transform = ax.transAxes)
ax.text(0.5, 1.0175, "github.com/datalib/libextract",
         horizontalalignment='center',
         fontsize=16,
         transform = ax.transAxes)
ax.text(0.5, 1.0075, "wikipedia.org/wiki/List_of_countries_by_life_expectancy",
         horizontalalignment='center',
         fontsize=12,
         transform = ax.transAxes)

leg = ax.legend(loc="upper right")
# Get the bounding box of the original legend
bb = leg.legendPatch.get_bbox().inverse_transformed(ax.transAxes)

# Change to location of the legend. 
xOffset = 1.0
newX0 = bb.x0
newX1 = bb.x1
bb.set_points([1,1.02])
leg.set_bbox_to_anchor(bb)
ax.figure
```
